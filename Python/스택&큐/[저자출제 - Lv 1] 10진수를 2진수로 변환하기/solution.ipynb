```python
'''
아이디어 1
bin 함수 사용
'''
def solution(N):
    
    return str(bin(N))[2:]

solution(12345)
```

```python
'''
아이디어 2
내가 직접 bin 함수 만들어 보기
'''
def solution(N):
    
    result = []
    # 2로 계속 나눠서 묶이 1될 때 까지 나머지 저장하기
    while N > 0:
        
        N,s = N//2,  N%2
        result.append(s)

    return "".join(map(str,result[::-1]))

solution(12345)
```

```python
# 테스트 케이스
input = 10
output = "1010"
```

```python
# 테스트 함수
def test(solution):

    # 비교하기
    if output == solution(input):
        print(f'예상 값={output}, 출력 값={solution(input)}')
        return "통과"
    else:
        print(f'예상 값={output}, 출력 값={solution(input)}')
        return "틀림"
```

```python
test(solution)
```
