```python
'''
아이디어 1
스택 사용하는 전형적인 문제  
파이썬은 리스트로 구현가능(pop, append)
'''
def solution(s):
    # 임시 스택: 임시 스택에 값이 있으면 False 없으면 True
    stack = []
    # s의 첫번째가 pop()로 나오게 하기위해 역순 복사
    s = list(s)[::-1]
    
    # s 스택이 비어있을 때까지 반복
    while s:
        
        # 일단 하나 뽑기
        a = s.pop()
        # 순서 중요! 비여있는데 인덱스를 접근하려고 하면 오류뜸
        # 그래서 stack이 비어있다면 바로 else로 넘어감
        # ')' 로 체크해주는 이유는 )( 를 거르기 위한 조건
        # 두번째 조건 까지는 () 나 )( 전부 옳다고 보니까
        if stack and stack[-1] != a and a==')': 
            stack.pop()
        else:
            stack.append(a)

    # stack에 값이 있으면 False 없으면 True
    if stack:
        return False
    else:
        return True
    
solution(')(')
```

```python
# 테스트 케이스
input = "()()"
output = True
```

```python
# 테스트 함수
def test(solution):

    # 비교하기
    if output == solution(input):
        print(f'예상 값={output}, 출력 값={solution(input)}')
        return "통과"
    else:
        print(f'예상 값={output}, 출력 값={solution(input)}')
        return "틀림"
```

```python
test(solution)
```
