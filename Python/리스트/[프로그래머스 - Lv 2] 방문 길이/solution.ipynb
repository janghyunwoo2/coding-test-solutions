```python
'''
아이디어 1
좌표평면 구현을 위한 0으로 초기화된 2차원 배열을 생성 (11*11)
지나갈 때 마다 1로 카운팅하기
이미 지나간 길(1로 카운팅 된 곳), 좌표평면 벗어나면(인덱스 아웃) 무시
'''
def solution(dirs):
    answer = 0

    # 좌표평면 구현(11x11)
    coor_pln = [[ 0 for _ in range(11)] for _ in range(11)]
    a = set()

    # 스타트 index(5,5)
    point_i = 5
    point_j = 5

    # 이동하기
    for v in dirs:
        # 뱡향 구현
        ''' 
        U = coor_pln[i-1][j]
        D = coor_pln[i+1][j]
        R = coor_pln[i][j+1]
        L = coor_pln[i][j-1]
        '''
        if v == 'U':
            if point_i-1 != -1:
                point_i-=1
                a.add(tuple(sorted(((point_i+1,point_j),(point_i,point_j))))) # (이전 좌표),(현제 좌표)
        elif v == 'D':
            if point_i+1 != 11:
                point_i+=1
                a.add(tuple(sorted(((point_i-1,point_j),(point_i,point_j)))))
        elif v == 'R':
            if point_j+1 != 11:
                point_j+=1
                a.add(tuple(sorted(((point_i,point_j-1),(point_i,point_j)))))
        elif v == 'L':
            if point_j-1 != -1:
                point_j-=1
                a.add(tuple(sorted(((point_i,point_j+1),(point_i,point_j)))))


        
    return len(a)

solution('RRRLLL')
```

```python
''' 
 답지..
'''

def solution(dirs):
    answer = 0

    # 1. 지나간 '길'을 저장할 집합 (set) 선언
    # (x1, y1)에서 (x2, y2)로 가는 길과 (x2, y2)에서 (x1, y1)로 가는 길을 동일하게 취급하기 위해
    # 항상 '작은 좌표 -> 큰 좌표' 순서로 정규화하여 저장할 거예요.
    # 이렇게 하면 set의 자동 중복 제거 기능으로 왕복하는 길을 한 번만 계산하게 됩니다.
    visited_paths = set()

    # 2. 초기 로봇의 현재 좌표 (정중앙 5,5)
    # 파이썬 리스트 인덱스나 게임판에서는 보통 0부터 시작하지만,
    # 여기서는 문제의 좌표계가 -5 ~ 5까지 총 11*11 칸이므로,
    # 중앙을 (0,0)으로 두고 실제 인덱스를 0~10으로 변환하는게 편리해요.
    # -5 ~ 5 범위의 좌표를 0 ~ 10 범위의 인덱스로 변환: +5 해주기
    # 예를 들어, (0,0)은 (5,5)가 되고, (-5,-5)는 (0,0), (5,5)는 (10,10)이 됩니다.
    current_x, current_y = 5, 5 # (0,0)에 해당

    # 3. 각 방향에 따른 x, y 좌표 변화량 딕셔너리
    # 코드를 더 깔끔하게 만들 수 있는 팁이에요!
    move_directions = {
        'U': (0, 1),   # Up: y 증가
        'D': (0, -1),  # Down: y 감소
        'R': (1, 0),   # Right: x 증가
        'L': (-1, 0)   # Left: x 감소
    }

    # 4. 이동 명령 실행
    for cmd in dirs:
        # 이동하기 전의 현재 좌표를 저장해둡니다.
        # 이 좌표가 다음 이동에서 '이전 좌표'가 될 거예요.
        prev_x, prev_y = current_x, current_y

        # 명령에 따라 좌표 변화량 가져오기
        dx, dy = move_directions[cmd]

        # 새로운 예측 좌표 계산
        next_x, next_y = current_x + dx, current_y + dy

        # 5. 좌표평면을 벗어나는지 확인 (0~10 범위 유지)
        if not (0 <= next_x <= 10 and 0 <= next_y <= 10):
            # 범위를 벗어나면 이동하지 않고 다음 명령으로 넘어갑니다.
            continue

        # 6. 유효한 이동인 경우, 현재 로봇의 위치 업데이트
        current_x, current_y = next_x, next_y

        # 7. '길' 기록하기
        # (이전 좌표, 현재 좌표) 형태의 길을 만듭니다.
        # 문제의 핵심: (A->B)와 (B->A)를 같은 길로 취급하기 위해 정규화합니다.
        # 두 좌표 튜플 중 사전순으로 먼저 오는 것을 첫 번째로, 뒤에 오는 것을 두 번째로 둡니다.
        # 예를 들어, ((5,5), (5,6)) 과 ((5,6), (5,5)) 중 작은 것부터 나열:
        # min((5,5), (5,6)) = (5,5)
        # max((5,5), (5,6)) = (5,6)
        # -> 최종적으로 ( (5,5), (5,6) ) 으로 기록됩니다.

        # 8. 정규화된 길을 set에 추가 (set이 자동으로 중복을 제거해줍니다)
        visited_paths.add(tuple(sorted(((prev_x, prev_y), (current_x, current_y)))))

    # 9. set에 저장된 길의 개수가 새로운 길의 총 개수입니다.
    answer = len(visited_paths)
    return answer
```
