```python
'''
https://school.programmers.co.kr/learn/courses/30/lessons/181930
주사위 게임 2
'''

def solution(a, b, c):
    a = [a,b,c]
    lens = len(set(a))

    if lens==3:
        return sum(a)
    elif lens ==2:
        return sum(a)*sum(map(lambda x:x**2,a))
    elif lens ==1:
        return sum(a)*sum(map(lambda x:x**2,a))*sum(map(lambda x:x**3,a))

    return lens

    
solution(2,6,1)
solution(5,3,3)
solution(4,4,4)

```
