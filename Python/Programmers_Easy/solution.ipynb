```python
'''
https://school.programmers.co.kr/learn/courses/30/lessons/120849
모음 제거

'''
def solution(my_string:str):
    
    #mu = ['a', 'e', 'i', 'o', 'u']
    mu = 'aeiou'

    return ''.join(i for i in my_string if i not in mu)

    
solution("bus")
```
